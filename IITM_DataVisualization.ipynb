Module 6: Assignment 1 - Data Visualization


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/cars-1.csv")
df.head()
```

```python
plt.plot(df['model'], df['hp'])  # Plot the chart
plt.xlabel("Models of the cars")  # add X-axis label
plt.ylabel("Horse-Power of Cars")  # add Y-axis label
plt.title("Model Names vs HorsePower")  # add title
plt.show() 
```

Module 6: Assignment 2 - Data Visualization


```python
df1 = pd.read_csv("/content/cars-2.csv")
df1.head()
```

```python
df1.boxplot(column = 'gear',by = ['carb'],color = 'blue')
plt.xlabel("Number of carburetors")  # add X-axis label
plt.ylabel("Number of forward gears")  # add Y-axis label
plt.title("carbs vs gear")  # add title
plt.show() 
```

Module 6: Assignment 3 - Data Visualization

```python
df2 = pd.read_csv("/content/cars-3.csv")
df2.head()
```

```python
plt.hist(df2['wt'],histtype='stepfilled',color='forestgreen',bins=30)
plt.xlabel('weight of the cars')
plt.ylabel("Count")
plt.title('Histogram for the weight values')
```
